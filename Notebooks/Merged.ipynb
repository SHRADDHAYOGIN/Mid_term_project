```python
# imports numpy and pandas
import pandas as pd
import numpy as np
```

### Flights Dataset
---

```python
# Data Path
Data = "C:/Users/patel/lighthouse-lab-allprojects/Mid-Project-Data/Flights_jan_clean1.csv"
```

```python
Flights = pd.read_csv(Data, low_memory=False)
```

```python
Flights.columns
```

```python
Flights['fl_date'] = pd.to_datetime(Flights['fl_date'])
```

```python
Flights['day'] = Flights['fl_date'].dt.day
Flights['month'] = Flights['fl_date'].dt.month
Flights['year'] = Flights['fl_date'].dt.year
```

```python
# Data Path
Data = 'C:/Users/patel/lighthouse-lab-allprojects/Mid-Project-Data/passenger_clean.csv'
```

```python
Passenger = pd.read_csv(Data, low_memory=False)
```

```python
Passenger.columns
```

```python
# Data Path
Data = 'C:/Users/patel/lighthouse-lab-allprojects/Mid-Project-Data/fuel_consumption_clean.csv'
```

```python
Fuel_consumption = pd.read_csv(Data, low_memory=False)
```

### Fuel and Passengers
---

```python
common_columns = set(Flights.columns).intersection(Passenger.columns)
print(common_columns)
```

```python
common_columns = set(Passenger.columns).intersection(Fuel_consumption.columns)
print(common_columns)
```

```python
Flights['op_unique_carrier'].unique()
```

```python
Passenger['unique_carrier'].unique()
```

```python
Fuel_consumption['unique_carrier'].unique()
```

```python
merged_df1 = pd.merge(Flights, Passenger, left_on=['dest_airport_id', 'origin_airport_id', 'month', 'year','op_unique_carrier'],
                     right_on=['dest_airport_id', 'origin_airport_id', 'month', 'year','unique_carrier'])
```

```python
merged_df1
```

### Final dataset
---

```python
merged_df = pd.merge(merged_df1, Fuel_consumption, on=['airline_id', 'carrier_name', 'carrier', 'unique_carrier', 'month', 'year'])
```

```python
merged_df
```

```python
# Saving the DataFrame as a CSV file
merged_df.to_csv('C:/Users/patel/lighthouse-lab-allprojects/Mid-Project-Data/merged1.csv', index=False)
```
