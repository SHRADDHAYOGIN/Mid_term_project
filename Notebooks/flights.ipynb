```python
# Checking for duplicate rows
duplicate_rows = df.duplicated()

# Printing the duplicate rows
print(df[duplicate_rows])
```

```python
df[['arr_delay','dep_delay','carrier_delay', 'weather_delay', 'nas_delay', 'security_delay',
       'late_aircraft_delay']]
```

```python
# Replacing null values with zeros in column1 and column2
df[['arr_delay','dep_delay','carrier_delay', 'weather_delay', 'nas_delay', 'security_delay',
       'late_aircraft_delay']] = df[['arr_delay','dep_delay','carrier_delay', 'weather_delay', 'nas_delay', 'security_delay',
       'late_aircraft_delay']].fillna(0)
```

```python
df[['arr_delay','dep_delay','carrier_delay', 'weather_delay', 'nas_delay', 'security_delay',
       'late_aircraft_delay']]
```

```python
df['cancellation_code'].unique()
```

```python
# filling missing values
df['cancellation_code'].fillna('NC', inplace=True)
```

```python
df['cancellation_code'].unique()
```

```python
df[['crs_elapsed_time','actual_elapsed_time']]
```

```python
# Creating a new column by 'crs_elapsed_time','actual_elapsed_time'
df['Total_delay'] = np.where(df['actual_elapsed_time'].isnull() | df['crs_elapsed_time'].isnull(), np.nan, df['actual_elapsed_time'] - df['crs_elapsed_time'])
```

```python
df[['crs_arr_time', 'arr_time', 'arr_delay','dep_delay','crs_dep_time' , 'dep_time']]
```

```python
# we can drop 'crs_elapsed_time','actual_elapsed_time','crs_arr_time', 'arr_time', 'crs_dep_time' , 'dep_time'
df.drop(['crs_elapsed_time','actual_elapsed_time','crs_dep_time' , 'dep_time'], axis=1, inplace=True)
```

```python
#Drop almost null columns
df.drop(['first_dep_time', 'total_add_gtime','longest_add_gtime', 'no_name'], axis=1, inplace=True)
```

```python
selected_rows = (df['air_time'].isnull()) & (df['cancellation_code'] != 'NC')
```

```python
# Replacing null values with 0 in the main DataFrame
df.loc[selected_rows, 'air_time'] = df.loc[selected_rows, 'air_time'].fillna(0)
```

```python
df[(df['air_time'].isnull()) & (df['cancellation_code'] != 'NC')]
```

```python
df.info()
```

```python
df['Total_delay'].max()
```

```python
df[(df['Total_delay'].isnull()) & (df['cancellation_code'] != 'NC')]
```

```python
selected_rows = (df['Total_delay'].isnull()) & (df['cancellation_code'] != 'NC')
```

```python
# Replacing null values with 0 in the main DataFrame
df.loc[selected_rows, 'Total_delay'] = df.loc[selected_rows, 'Total_delay'].fillna(1000)
```

```python
df[(df['Total_delay'].isnull()) & (df['cancellation_code'] != 'NC')]
```

```python
# Removing rows with null values
df= df.dropna()
```
